```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import PillowWriter
```

```python
d = 0.4 #meters
```

```python
g = 9.8 #m/s^2
a_max = 2 #m/s^2
v_max = 0.5 #m/s
```

```python
T_v = d*math.pi/(2*v_max)
T_a = math.pi*math.sqrt(d/(2*a_max))
T = 0
if T_v > T_a:
    print('T_v =',T_v)
    T = T_v
else:
    print('T_a =',T_a)
    T = T_a
```

```python
t = 0
interval = .005
x = []
y_p = []
y_v = []
y_a = []

fig, axs = plt.subplots(3)
axs[0].set_xlim(0,T)
axs[0].set_ylim(-.005,d+.005)
axs[1].set_xlim(0,T)
axs[1].set_ylim(0,v_max + .05)
axs[2].set_xlim(0,T)
axs[2].set_ylim(-a_max-.1,a_max+.1)
writer = PillowWriter(fps = 25)
with writer.saving(fig,'pva7.gif',100):
    while (t <= T):
        x.append(t)
        y_p.append((-d)/2*math.cos(math.pi/T*t) + d/2)
        y_v.append(d*math.pi/(2*T)*math.sin(math.pi/T*t))
        y_a.append(d*math.pi**2/(2*T**2)*math.cos(math.pi/T*t))
        t += interval
        axs[0].plot(x,y_p,'-k')
        axs[1].plot(x,y_v,'-b')
        axs[2].plot(x,y_a,'-r')
        writer.grab_frame()
    #plt.close()
```
